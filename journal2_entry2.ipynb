```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

def calculate_sin(val):
    return np.sin(val)

def calculate_cos(val):
    return np.cos(val)
```

```python
f, axarr = plt.subplots(1, 2, figsize=(10, 4))

axarr[0].plot(x, calculate_sin(x))
axarr[0].set_title('sin(x)')
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('Amplitude')

axarr[1].plot(x, calculate_cos(x), color='orange')
axarr[1].set_title('cos(x)')
axarr[1].set_xlabel('x')

plt.tight_layout()

plt.savefig('trig_panels.pdf')

plt.show()
```
