# Sine Function
This function returns sin(x) and and tabulates the values over a specific range.

```python
import numpy as np

def sin_function(x):
    """Returns the sine of x."""
    return np.sin(x)
```

# Cosine Function
This function returns cos(x) and and tabulates the values over a specific range.

```python
def cos_function(x):
    """Returns the cosine of x."""
    return np.cos(x)
```

# Data Generation and Tabulation
This will generate 1,000 values of $x$ between 0 and 2. 
We will use these functions to calculate the corresponding sine and cosine values.

```python
x_values = np.linspace(0, 2, 1000)

sin_values = sin_function(x_values)
cos_values = cos_function(x_values)
```

# Results Table
This cell prints the first 10 values of our results in a formatted table.

```python
print(f"{'x':>10} | {'sin(x)':>10} | {'cos(x)':>10}")
print("-" * 35)

for i in range(10):
    print(f"{x_values[i]:10.4f} | {sin_values[i]:10.4f} | {cos_values[i]:10.4f}")
```
