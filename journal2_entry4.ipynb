```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.gamma(2.0 , 1.0, 1000)
```

```python
plt.hist(x, bins=100, edgecolor='white', linewidth=0.3)

plt.xlabel('Value')
plt.ylabel('Frequency (Count)')
plt.title('Gamma Distribution (Shape=2.0)')

plt.savefig('gamma_histogram.pdf')

plt.show()
```
